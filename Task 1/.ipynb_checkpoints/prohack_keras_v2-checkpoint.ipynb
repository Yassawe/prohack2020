```python
class prohack:
    def __init__(self, hidden_layers):
        self.layers=hidden_layers
        self.L=len(hidden_layers)
        self.optimizer=None
        self.model=None
    
    def create_model(self, input_shape, activations, dropout, normalize, dropout_rate, norm_momentum, epsilon):
        model = Sequential()
        model.add(Dense(self.layers[0], activation=activations[0], kernel_regularizer=regularizers.l1_l2(l1=1e-4, l2=1e-3), input_shape=input_shape))
        for i in range(1, self.L):
            if normalize:
                model.add(BatchNormalization(axis=-1, momentum=norm_momentum, epsilon=epsilon))
            if dropout:
                model.add(Dropout(rate=dropout_rate))
            model.add(Dense(self.layers[i], activation=activations[i], kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4)))
        self.model = model
            
    def create_optimizer(self,learning_rate, decay_rate, decay_base, beta1, beta2, epsilon):
        scheduled_lr=ExponentialDecay(learning_rate, decay_base, decay_rate)
        self.optimizer = Adam(scheduled_lr, beta1, beta2, epsilon)
    
    def RMSE_loss(self, y, yhat):
        return K.sqrt(K.mean(K.square(yhat - y)))
    
    def train(self, X_train, Y_train, batch_size, loss, epochs):
        if loss=="rmse":
            loss=self.RMSE_loss
        self.model.compile(loss=loss, optimizer=self.optimizer, metrics=[self.RMSE_loss])
        self.model.fit(X_train, Y_train, batch_size, epochs, verbose=2)
    
    def evaluate(self, X,Y):
        results = self.model.evaluate(X, Y, batch_size=X.shape[0])
        return results
    
    def predict(self, X):
        prediction=self.model.predict(X)
        return prediction
        
```

```python
#PARAMETERS OF THE NETWORK ARCHITECTURE
layers=[128, 320, 320, 50, 20, 1] #каждый элемент массива это кол-во нейронов в лейере, длина массива - кол-во лейеров в сети (не считая инпута)
activations=["sigmoid", "sigmoid","sigmoid","sigmoid","sigmoid","sigmoid"]

"""
в keras есть следующие activation functions:
sigmoid
relu
selu
elu
tanh
softplus
softsign
"""

#PARAMETERS OF THE LAYER MODE
dropout=True
normalize=True
dropout_rate=0.2
norm_momentum=0.5
epsilon=1e-7

#PARAMETERS OF OPTIMIZATION
learning_rate=0.001
decay_rate=0.9 
decay_base=2000
"""
new_learning_rate=learning_rate*decay_rate*(current_step/decay_base)

i.e чем меньше decay_rate тем сильнее learning rate уменьшиться
чем меньше decay_base тем чаще он это будет делать
поставьте decay_rate=1 чтобы отключить learning rate decay
"""

beta1=0.9
beta2=0.99


#PARAMETERS OF LEARNING
batch_size=256 #поставьте 2532 чтобы тренировать модель по целому сету, а не по minibatch
epochs=10000
loss = "rmse"

"""
в keras есть следующие loss functions для регрессии которые нам подходят

mean_absolute_error
mean_squared_error
mean_absolute_percentage_error
mean_squared_logarithmic_error
huber_loss
log_cosh
___________________
можно поставить rmse

"""


input_shape=(77,)
```

```python
network=prohack(layers)
network.create_model(input_shape, activations, dropout, normalize, dropout_rate, norm_momentum, epsilon)
network.create_optimizer(learning_rate, decay_rate, decay_base, beta1, beta2, epsilon)
```

```python
network.train(X_train,Y_train,batch_size, loss, epochs)
```

```python
print("training set error: ")
train_error=network.evaluate(X_train, Y_train)
print("test set error: ")
test_error=network.evaluate(X_test, Y_test)
```
