```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer  # noqa
from sklearn.impute import IterativeImputer
from sklearn.linear_model import BayesianRidge
```

```python
data=pd.read_csv("original datasets/train.csv")
```

```python
imputer=IterativeImputer(estimator=BayesianRidge(), sample_posterior=True, imputation_order="random")
```

```python
data.head()
```

```python
galaxies=data.galaxy.unique().tolist()
```

```python
frames=[]
for galaxy in galaxies:
    df = data.loc[data.galaxy==galaxy, data.columns!= 'galaxy']
    columns=df.columns
    array=df.to_numpy()
    new_array=imputer.fit_transform(array)
    df2 = pd.DataFrame(new_array, columns=columns)
    df2.insert(1, 'galaxy', galaxy)
    frames.append(df2)

transformed_data=pd.concat(frames)
    
```
