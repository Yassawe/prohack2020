```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
import pandas as pd
import seaborn as sns
#pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
train_set=pd.read_csv("original datasets/train.csv")
test_set=pd.read_csv("original datasets/test.csv")
```

```python
galaxies=train_set.galaxy.unique().tolist()

for galaxy in galaxies:
    subset = train_set.loc[train_set.galaxy==galaxy, :]
    index = pd.Index(subset.loc[:,"galactic year"])
    
    values=subset.loc[:, "y"].values
    ser=pd.Series(values, index)
    ser.plot()
    
```

```python
for i in range(2,79):
    Figure()
    print(train_set.columns[i])
    train_set.iloc[:, i].plot.hist(bins=100)
    test_set.iloc[:,i].plot.hist(bins=100)
    plt.show()
    
```

```python
trna=train_set.isnull()
tsna=test_set.isnull()
```
